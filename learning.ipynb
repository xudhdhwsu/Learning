```python
import pandas as pd;
import numpy as np;

data = pd.read_csv("Learning\\src\\main\\resources\\case1.csv");

wt = data.iloc[: ,1];
volume = data.iloc[: ,2];
```

```python
def f(k, b, wt):
    return k * wt + b;
```

```python
def loss(k ,b , wt, volume):
    loss = 0;
    for i in range(len(wt)):
        y = f(k, b, wt[i]);
        loss = loss + (y - volume[i]) ** 2;
    return loss / len(wt);

""" 矩阵运算 """

def loss(k, b):
    return np.mean((f(k, b, wt) - volume) ** 2);
```

```python
def grandient(k, b):
    grandientk = [loss(k + 0.0001, b) - loss(k, b)] / 0.0001;
    grandientb = [loss(k, b + 0.0001) - loss(k, b)] / 0.0001;
    return grandientk, grandientb;
```
